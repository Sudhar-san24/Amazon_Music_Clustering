```python
# ======================
# 1. Import Libraries
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# ======================
# 2. Load Dataset
# ======================
df = pd.read_csv('D:\Amazon_Music_Clustering\Dataset\single_genre_artists.csv')

# Quick checks
print(df.info())
print(df.describe())
print(df.describe(include=object))
print(df.isnull().sum())

# ======================
# 3. Reference Columns (not used for clustering)
# ======================
reference_columns = ["id_songs", "name_song", "id_artists", "name_artists", 
                     "release_date", "genres", "followers"]
reference_df = df[reference_columns]

# ======================
# 4. Outlier Removal (Z-score method)
# ======================
zscore_cols = [
    "popularity_songs", "duration_ms", "danceability", "energy",
    "loudness", "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo", "popularity_artists"
]
z_scores = df[zscore_cols].apply(zscore)
df_no_outliers = df[(abs(z_scores) < 3).all(axis=1)]

# ======================
# 5. Feature Selection
# ======================
num_features = [
    "popularity_songs", "duration_ms", "danceability", "energy",
    "loudness", "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo", "popularity_artists"
]
cat_features = ["explicit", "mode", "key", "time_signature"]

# Prepare numeric and categorical data
X_num = df_no_outliers[num_features]
X_cat = pd.get_dummies(df_no_outliers[cat_features], drop_first=True)

# Final feature matrix
X = pd.concat([X_num, X_cat], axis=1)

# ======================
# 6. Scaling
# ======================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ======================
# 7. K-Means → Elbow Method
# ======================
inertia = []
K = range(2, 15)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()

# ======================
# 8. Final KMeans Model
# ======================
kmeans = KMeans(n_clusters=10, random_state=42)   # <-- update after elbow
kmeans.fit(X_scaled)

labels = kmeans.labels_
X["Cluster"] = labels

# ======================
# 9. Cluster Visualization
# ======================
# Simple 2D (first 2 features)
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.title("KMeans Clustering (raw features)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# PCA 2D Projection
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.title("KMeans Clustering with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()

# ======================
# 10. Silhouette Scores
# ======================
print(f"Silhouette Score (k=10): {silhouette_score(X_scaled, labels):.4f}")

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    print(f"k={k} → Silhouette Score: {score:.4f}")

# ======================
# 11. DBSCAN
# ======================
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

# Handle noise points (-1 label)
mask = db_labels != -1
if mask.sum() > 0:
    print("DBSCAN Silhouette Score:", silhouette_score(X_scaled[mask], db_labels[mask]))

# PCA Projection for DBSCAN
X_pca = PCA(n_components=2).fit_transform(X_scaled)
plt.scatter(X_pca[mask, 0], X_pca[mask, 1], c=db_labels[mask], cmap='tab10', label='Clusters')
plt.scatter(X_pca[~mask, 0], X_pca[~mask, 1], c='gray', label='Noise (-1)', alpha=0.5)
plt.title('DBSCAN clustering results with noise')
plt.legend()
plt.show()
```
