```python
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
# ======================
# 4. Outlier Removal (Z-score method)
# ======================
zscore_cols = [
    "popularity_songs", "duration_ms", "danceability", "energy",
    "loudness", "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo", "popularity_artists"
]
z_scores = df[zscore_cols].apply(zscore)
df_no_outliers = df[(abs(z_scores) < 3).all(axis=1)]

# ======================
# 5. Feature Selection
# ======================
num_features = [
    "popularity_songs", "duration_ms", "danceability", "energy",
    "loudness", "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo", "popularity_artists"
]
cat_features = ["explicit", "mode", "key", "time_signature"]

# Prepare numeric and categorical data
X_num = df_no_outliers[num_features]
X_cat = pd.get_dummies(df_no_outliers[cat_features], drop_first=True)

# Final feature matrix
X = pd.concat([X_num, X_cat], axis=1)

# ======================
# 6. Scaling
# ======================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score 
#11. DBSCAN
# ======================
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

# Handle noise points (-1 label)
mask = db_labels != -1
if mask.sum() > 0:
    print("DBSCAN Silhouette Score:", silhouette_score(X_scaled[mask], db_labels[mask]))



```

```python
# PCA Projection for DBSCAN
X_pca = PCA(n_components=2).fit_transform(X_scaled)
plt.scatter(X_pca[mask, 0], X_pca[mask, 1], c=db_labels[mask], cmap='tab10', label='Clusters')
plt.scatter(X_pca[~mask, 0], X_pca[~mask, 1], c='red', label='Noise (-1)', alpha=0.5)
plt.title('DBSCAN clustering results with noise')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Select audio features
features = [
    "danceability", "energy", "loudness", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo", "duration_ms"
]

X = df[features]  # Use your cleaned DataFrame here
X
```

```python
# Apply StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=features)
X_scaled_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (SSE)")
plt.title("Elbow Method for Optimal k")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Apply KMeans with optimal k
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to your original DataFrame
df["KMeans_Cluster"] = kmeans_labels
```

```python
from sklearn.metrics import silhouette_score

kmeans_score = silhouette_score(X_scaled, kmeans_labels)
print("K-Means Silhouette Score:", round(kmeans_score, 3))
```

```python
from sklearn.cluster import KMeans

kmeans_pca = KMeans(n_clusters=3, random_state=42)  # Use best k from earlier
kmeans_pca_labels = kmeans_pca.fit_predict(X_pca)
```

```python
# 7. K-Means 
from sklearn.cluster import KMeans, DBSCAN
inertia = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# 8. Final KMeans Model

kmeans = KMeans(n_clusters=5, random_state=42)   # <-- update after elbow
kmeans.fit(X_scaled)
```

```python
labels = kmeans.labels_
X["Cluster"] = labels
```

```python
# 9. Cluster Visualization
# ======================
# Simple 2D (first 2 features)
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.title("KMeans Clustering (raw features)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
